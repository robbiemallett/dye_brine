```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import string
from matplotlib import gridspec
```

```python
letters = string.ascii_lowercase

photos_dir = '../photos/field_photos'

fnames = {'DP1 9th Dec':f'{photos_dir}/DP0_9th/20211209_112355_crop.jpg',
          'DP1 11th Dec':f'{photos_dir}/DP0_11th/20211211_114123.jpg',
          'DP1 12th Dec':f'{photos_dir}/DP0_12th/20211212_112145.jpg',
          'DP2 11th Dec':f'{photos_dir}/DP1_11th/20211211_121404.jpg',
          'DP2 12th Dec':f'{photos_dir}/DP1_12th/20211212_113557.jpg',
          'CP 13th Dec':f'{photos_dir}/LP1_13th/20211213_143951_crop.jpg',
         }
          

fig, axs = plt.subplots(3,2,figsize=(5,6))

for counter, (title, fname, ax, letter) in enumerate(zip(fnames.keys(),fnames.values(),
                                                         axs.reshape(-1), letters)):

    i = Image.open(fname)
    iar = np.array(i)
    
    if counter in [2]:    iar = np.flipud(iar)
    
    ax.imshow(iar)
    
    ax.set_title(title,pad=-2)
    

    ax.annotate(f'({letter})', fontsize='x-large', xy=(0.02,0.98),va='top',xycoords='axes fraction')
    
    ax.set_yticks([])
    ax.set_xticks([])
    
plt.subplots_adjust(wspace=0.05)

# plt.savefig('../Paper/Figures/Fig_3.jpg',dpi=500,bbox_inches='tight')
plt.savefig('/home/robbie/Dropbox/Apps/Overleaf/rothera_dye_brine_v2/figs/fieldresults_photos.png',dpi=500,bbox_inches='tight')
```

```python
letters = string.ascii_lowercase

photos_dir = '/home/robbie/Dropbox/dye_brine/photos'

fnames = {'x': f'{photos_dir}/strat_layer_cropped.jpg',
          'DP1 Setup 4th Dec': f'{photos_dir}/brine_rejection_tube.jpg',
          'DP2 Setup 9th Dec': f'{photos_dir}/disect_example_cropped.jpg',
            
         }
          

# fig, axs = plt.subplots(1,2,figsize=(8,5))

fig = plt.figure(constrained_layout=True,figsize=(6,5.8))
spec2 = gridspec.GridSpec(ncols=2, nrows=2, figure=fig,height_ratios=[10,11.1],width_ratios=(4,9),wspace=0.01)
ax3 = fig.add_subplot(spec2[0, :])
ax1 = fig.add_subplot(spec2[1, 0])
ax2 = fig.add_subplot(spec2[1, 1])




for counter, (fname, ax, letter) in enumerate(zip(fnames.values(),
                                                         (ax3,ax1,ax2), letters)):

    i = Image.open(fname)
    iar = np.array(i)
    
    if 'disect' in fname:
        iar = np.flipud(iar)
    
    if counter in [2]:    iar = np.flipud(iar)
    
    ax.imshow(iar)
    
    if letter == 'a':t = ax.annotate(f'({letter})', fontsize='x-large', xy=(0.988,0.97),va='top',ha='right',xycoords='axes fraction')
    elif letter == 'b':t = ax.annotate(f'({letter})', fontsize='x-large', xy=(0.96,0.975),va='top',ha='right',xycoords='axes fraction')
    elif letter == 'c':t = ax.annotate(f'({letter})', fontsize='x-large', xy=(0.984,0.976),va='top',ha='right',xycoords='axes fraction')
    
    t.set_bbox(dict(facecolor='white', alpha=0.8, edgecolor='white'))
    
    ax.set_yticks([])
    ax.set_xticks([])

# plt.subplots_adjust(wspace=0.02)

# plt.savefig('../Paper/Figures/fieldsetup.jpg',dpi=500,bbox_inches='tight')
plt.savefig('/home/robbie/Dropbox/Apps/Overleaf/rothera_dye_brine_v2/figs/labsetup.png',dpi=500,bbox_inches='tight')


```
