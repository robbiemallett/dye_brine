```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import datetime
```

```python
df = pd.read_csv('../data/temperatureprobe_.csv')
df['TS'] = pd.to_datetime(df['Timestamp'])

skipper = 60
xticks = [datetime.datetime(2023,10,5,0,0,0) + datetime.timedelta(days=int(x)) for x in np.arange(11)]


plt.plot(df['TS'][::skipper],df['Temperature (C)'][::skipper])

# df_exp = df[df['TS']>datetime.datetime(2023,10,6,10)]
# df_exp = df_exp[df_exp['TS']<datetime.datetime(2023,10,15,0)]
# plt.plot(df_exp['TS'][::skipper],df_exp['Temperature (C)'][::skipper])


plt.xticks(xticks)

plt.xlim(datetime.datetime(2023,10,5),datetime.datetime(2023,10,16))
plt.ylim(-6.5,-4.8)

plt.ylabel(r'Temperature ($\degree$C)', fontsize='x-large')

plt.xticks(rotation=90)

df_exp = df[df['TS']>datetime.datetime(2023,10,6,10)]
df_exp = df_exp[df_exp['TS']<datetime.datetime(2023,10,15,0)]
ts = df_exp['Temperature (C)']


plt.axhline(np.nanmean(ts),color='k',ls='--')
# plt.axhline(np.nanmean(ts)+np.nanstd(ts),color='k',ls=':')
# plt.axhline(np.nanmean(ts)-np.nanstd(ts),color='k',ls=':')

plt.savefig('/home/robbie/Dropbox/Apps/Overleaf/rothera_dye_brine_v2/supp_figs/temp_fluctuations.png',bbox_inches='tight',dpi=500)

plt.show()
```
